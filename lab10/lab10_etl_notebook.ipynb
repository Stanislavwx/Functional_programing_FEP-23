
## 1) Огляньмо зразки даних


```

base = Path("/mnt/data")
print((base / "lab10_sample.csv").read_text(encoding="utf-8"))
print((base / "lab10_sample.json").read_text(encoding="utf-8"))
```


## 2) Проганяємо ядро над CSV


```

rows = list(core_pipeline(read_csv(base / "lab10_sample.csv")))
rows[:3]
```


## 3) Запис у JSON і назад


```

out_json = base / "lab10_out.json"
write_json(out_json, rows)
print("JSON saved to:", out_json)
print(out_json.read_text(encoding="utf-8")[:400], "...")
```


## 4) End-to-end: JSON → CSV


```

out_csv = base / "lab10_out.csv"
etl_json_to_csv(base / "lab10_sample.json", out_csv)
print("CSV saved to:", out_csv)
print(out_csv.read_text(encoding="utf-8"))
```
